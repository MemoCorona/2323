```python
path = "datasets/Cartwheeldata.csv"
df = pd.read_csv(path)
df.head(2)
```

```python
import seaborn as sns
sns.lmplot(x = "Wingspan",
          y = "CWDistance",
          data = df,
          fit_reg = True,#Linea de regresión
          hue = "Gender") #Hue para definir 
plt.title("Scatter plot of Wingspan vs CWDistance")
plt.show()

```

```python
sns.swarmplot(x = "Gender",
             y = "CWDistance",
             data = df
             )
plt.title("Gender vs CWDistance")
plt.show()
```

```python
sns.boxplot(data = df.loc[df["Gender"]=="M",
           ["Age", "Height","Wingspan", "CWDistance"]]
           )
plt.show()
```

```python
#BOX PLOT DOS FIGURAS
fig = plt.figure()

ax1 = fig.add_subplot(1, 3,1) # 1 renglon 2 columnas y posición 1 
sns.boxplot(data = df.loc[df["Gender"] == "M",
                          ["Score"]
                         ]
           )
ax2 = fig.add_subplot(1, 3,3) # 1 renglon 2 columnas y posición 1 
sns.boxplot(data = df.loc[df["Gender"] == "F",
                          ["Score"]
                         ]
           )
ax1.title.set_text("Male score")
ax2.title.set_text("Female score")
plt.show()
```

```python
sns.distplot(df.CWDistance)
plt.show()
```

```python
sns.countplot(x ="Gender",
             data = df)
plt.xticks(rotation = 45)
plt.yticks(rotation = 45)
plt.show()
```
